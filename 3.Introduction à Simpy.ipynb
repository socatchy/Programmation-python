a)À l'aide de Sympy calculons la fonction $g:\mathbb{R}\to\mathbb{R}$ définie en gardant les cent premiers termes de la série: $$
g(x) = \sum_{k=1}^{100} \frac{\sin(k^2 x)}{k^5} \,,
$$

```python
x = sp.Symbol('x')
k = sp.Symbol('k', integer=True)
borne_inf = 1
borne_sup = 100

terme_general = sp.sin(k**2 * x) / k**5

g_x = sp.summation(terme_general, (k, borne_inf, borne_sup))

x_valeur = sp.pi / 2
valeur_g = g_x.subs(x, x_valeur).evalf()

print(f"Le premier terme (k=1) est : {terme_general.subs(k, 1)}")
print(f"Le dernier terme (k=100) est : {terme_general.subs(k, 100)}")
print("-" * 50)
print(f"g(x) est défini par la somme de 100 termes trigonométriques.")
print(f"Exemple de calcul : g(π/2) ≈ {valeur_g}")
```

```python
import matplotlib.pyplot as plt
g_x_sym = sp.summation(terme_general, (k, borne_inf, borne_sup))
g_numerical = sp.lambdify(x, g_x_sym, 'numpy')
x_vals = np.linspace(-2 * np.pi, 2 * np.pi, 500)
y_vals = g_numerical(x_vals)

plt.figure(figsize=(12, 6))
plt.plot(x_vals, y_vals, label=r'$g(x) = \sum_{k=1}^{100} \frac{\sin(k^2 x)}{k^5}$', color='dodgerblue')

plt.title(r'Représentation graphique de $g(x) = \sum_{k=1}^{100} \frac{\sin(k^2 x)}{k^5}$')
plt.xlabel(r'$x$')
plt.ylabel(r'$g(x)$')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

Fonction de Green du laplacien

Notre but sera ici de calculer entièrement et automatiquement la fonction de Green du laplacien dans $\mathbb{R}^3$, *i.e.* la solution satisfaisant:
$$
\Delta G(\boldsymbol{x}) = \delta(\boldsymbol{x}) \,,
$$
dans $\mathbb{R}^3$ où $\delta(\boldsymbol{x})$ est la distribution de Dirac.

Pour cela, nous introduisons les coordonnées sphériques $\boldsymbol{x}^\prime = (r,\theta,\varphi)$ avec $r>0$, $0 \leq \theta \leq \pi$, et $0 \leq \varphi < 2\pi$ caractérisées par:
$$
\begin{align*}
x_1 & =r\cos\varphi\sin\theta\\ 
x_2 & =r\sin\varphi\sin\theta\\ 
x_3 & =r\cos\theta \,.
\end{align*}
$$

a)Définissons une fonction `to_spherical(expr)` permettant de convertir en coordonnées sphériques une expression donnée en coordonnées cartésiennes.

```python
def to_spherical(expr):
    x, y, z = sp.symbols('x y z')
    r, theta, phi = sp.symbols('r theta phi')

    substitution_rules = [
        (x, r * sp.sin(theta) * sp.cos(phi)),
        (y, r * sp.sin(theta) * sp.sin(phi)),
        (z, r * sp.cos(theta))
    ]

    spherical_expr = expr.subs(substitution_rules).simplify()

    return spherical_expr
```

b)Définissons une fonction `to_cartesian(expr)` permettant de convertir en coordonnées cartésiennes une expression donnée en coordonnées sphériques.
Par simplicité, on traite uniquement le cas d'une expression `expr` invoquant les variables $r$ et $\theta$ mais pas $\varphi$.

```python
def to_cartesian(expr):
    r, theta, phi = sp.symbols('r theta phi')
    x, y, z = sp.symbols('x y z')
    
    r_cartesian = sp.sqrt(x**2 + y**2 + z**2)
    

    cos_theta_cartesian = z / r_cartesian
    sin_theta_cartesian = sp.sqrt(x**2 + y**2) / r_cartesian

    substitution_rules = {
        r: r_cartesian,
        sp.cos(theta): cos_theta_cartesian,
        sp.sin(theta): sin_theta_cartesian
    }
    
    cartesian_expr = expr.subs(substitution_rules).simplify()

    return cartesian_expr
```

c) Calculons les facteurs d'échelle des coordonnées sphériques: $$
h_i = \left\Vert \frac{\partial\boldsymbol{x}}{\partial x^\prime_i} \right\Vert.

d)Définissons une fonction `gradient(f)` permettant de calculer le gradient d'une fonction $f:\mathbb{R}^3\to\mathbb{R}$ en coordonnées sphériques: $$
\boldsymbol{\nabla}f = \left( \frac{1}{h_i}\frac{\partial f}{\partial x^\prime_i} \right)_{i=1}^3\,.
$$

```python
def gradient_spherical(f):
    r, theta, phi = sp.symbols('r theta phi', real=True, positive=True)

    h_r = 1
    h_theta = r
    h_phi = r * sp.sin(theta)

    grad_r = (1 / h_r) * sp.diff(f, r)
    grad_theta = (1 / h_theta) * sp.diff(f, theta)
    grad_phi = (1 / h_phi) * sp.diff(f, phi)

    gradient_vector = sp.Matrix([grad_r, grad_theta, grad_phi])

    return gradient_vector.simplify()
```

e)Faisons de même pour définir le laplacien en coordonnées sphériques: $$
\Delta f = \sum_{i=1}^3 \frac{1}{J} \frac{\partial}{\partial x^\prime_i} \left( \frac{J}{h_i^2} \frac{\partial f}{\partial x^\prime_i} \right) \quad \text{où} \quad J = \prod_{i=1}^3 h_i \,.
$$

```python
def laplacian_spherical(f):
    r, theta, phi = sp.symbols('r theta phi', real=True, positive=True)
    coords = [r, theta, phi]

    h_r = 1
    h_theta = r
    h_phi = r * sp.sin(theta)
    
    h_factors = [h_r, h_theta, h_phi]
    
    J = h_r * h_theta * h_phi
    
    laplacian_sum = 0
    
    for coord, h_i in zip(coords, h_factors):
        inner_term = (J / h_i**2) * sp.diff(f, coord)
        outer_term = sp.diff(inner_term, coord)
        
        laplacian_sum += outer_term

    laplacian_f = (1 / J) * laplacian_sum
    
    return laplacian_f.simplify()
```

f)Trouvons les solutions radiales (*i.e.* ne dépendant que de la variable $r$) de l'équation $\Delta G = 0$ dans $\mathbb{R}^3\setminus\{\boldsymbol{0}\}$.

```python
r, theta, phi = sp.symbols('r theta phi', real=True, positive=True)
G = sp.Function('G')(r)

h_r = 1
h_theta = r
h_phi = r * sp.sin(theta)
J = h_r * h_theta * h_phi

laplacian_G = (1 / J) * sp.diff((J / h_r**2) * sp.diff(G, r), r)

laplacian_G_simplified = laplacian_G.simplify()
equation = sp.Eq(laplacian_G_simplified, 0)


solution = sp.dsolve(equation, G)


print(f"Équation de Laplace pour les solutions radiales (ΔG = 0) :")
print(equation)
print("-" * 60)
print("Solution générale G(r) :")
sp.pretty_print(solution)
```

g)Déterminons les équations que doivent satisfaire les constantes d'intégration pour que la solution précédente satisfasse en coordonnées cartésiennes: $$
\lim_{|\boldsymbol{x}|\to\infty} G(\boldsymbol{x}) = 0 \quad \text{et} \quad \Delta G(\boldsymbol{x}) = \delta(\boldsymbol{x}) \,.
$$

Déjà, le fait que G doit s'annuler en l'infinie implique que $C_2=0$

La deuxième condition est l'équation de Poisson homogène avec une source ponctuelle (la distribution de Dirac $\delta(\boldsymbol{x})$) :$$\Delta G(\boldsymbol{x}) = \delta(\boldsymbol{x})$$Puisque nous avons la solution radiale $G(\boldsymbol{x}) = \frac{C_1}{|\boldsymbol{x}|}$, nous devons trouver la valeur de $C_1$ qui satisfait cette condition. Pour ce faire, nous intégrons l'équation de Poisson sur un volume $V$ contenant l'origine (la singularité de la fonction de Dirac).$$\iiint_V \Delta G(\boldsymbol{x}) \,dV = \iiint_V \delta(\boldsymbol{x}) \,dV$$

Par définition de la fonction de Dirac, l'intégrale sur tout volume $V$ contenant l'origine est $1$ :$$\iiint_V \delta(\boldsymbol{x}) \,dV = 1$$Intégrale du terme de gauche (Laplacien)Nous utilisons le Théorème de Green (ou Théorème de la Divergence) pour convertir l'intégrale volumique du Laplacien en une intégrale surfacique du gradient. Soit $S$ la surface qui délimite le volume $V$ (par exemple, une sphère de rayon $R$ centrée à l'origine) :$$\iiint_V \Delta G \,dV = \oiint_S \boldsymbol{\nabla} G \cdot \mathbf{n} \,dS$$Où $\mathbf{n}$ est le vecteur unitaire normal à la surface $S$, pointant vers l'extérieur.Pour une sphère de rayon $R$, $\mathbf{n} = \boldsymbol{\hat{r}}$. Le gradient de $G = C_1/r$ en coordonnées sphériques est (comme calculé dans la partie d) :$$\boldsymbol{\nabla} G = \frac{1}{h_r}\frac{\partial G}{\partial r} \boldsymbol{\hat{r}} = 1 \cdot \frac{d}{dr}\left(\frac{C_1}{r}\right) \boldsymbol{\hat{r}} = -\frac{C_1}{r^2} \boldsymbol{\hat{r}}$$Sur la surface $S$ (où $r=R$), $\boldsymbol{\nabla} G = -\frac{C_1}{R^2} \boldsymbol{\hat{r}}$.L'intégrale devient :$$\oiint_S \boldsymbol{\nabla} G \cdot \mathbf{n} \,dS = \oiint_S \left(-\frac{C_1}{R^2} \boldsymbol{\hat{r}}\right) \cdot \boldsymbol{\hat{r}} \,dS = -\frac{C_1}{R^2} \oiint_S \,dS$$L'intégrale $\oiint_S \,dS$ est simplement l'aire de la surface de la sphère, qui est $4\pi R^2$ :$$\oiint_S \boldsymbol{\nabla} G \cdot \mathbf{n} \,dS = -\frac{C_1}{R^2} (4\pi R^2) = -4\pi C_1$$ÉgalisationEn égalisant les deux côtés de l'équation de Poisson :$$\iiint_V \Delta G \,dV = \iiint_V \delta(\boldsymbol{x}) \,dV$$$$-4\pi C_1 = 1$$$$\mathbf{C_1 = -\frac{1}{4\pi}}$$


Notes : les intégrales curvilignes doubles ne sont pas passées, elles correspondent aux symboles en rouge.
