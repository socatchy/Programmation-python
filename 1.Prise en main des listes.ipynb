d) Comment déterminer le n-ieme terme de la suite de Fibonnaci ? Voici une possibilité de code, avec un exemple en prenant N=5: ( on essaye aussi N=2 pour voir s'il n'y pas de problème au valeur extrème) : 

```python
def fibonacci(N):
    list = [0 for i in range(N)]
    list[1] = 1
    for i in range(2, N):
        list[i] = list[i-1] + list[i-2]
    return list
print(fibonacci(5))
print(fibonacci(2))
```

e) Comment avoir la n-ième ligne du triangle de Pascal ? J'ai préféré la description directe de la N-ième ligne à l'aide des coefficients binomiaux plutôt que de construire des listes par récurrences. D'abord, j'ai construit la factoriel en récursif :

```python
def fact(N):
    if (N == 1 or N==0):
        return 1
    return N*fact(N-1)
print(fact(4))
```

On donne maintenant la fonction qui retourne la N-ieme ligne du triangle de Pascal : 
1. Ce n'est pas forcement la méthode la plus efficace en terme de complexité 
2. Le rôle du N = n-1 est d'assurer que pascal(N) donne bien la N-ième ligne et non la (N+1)ième
3. On calcul au départ fact(N) pour eviter de le faire a chaque itération, on aurait pu aussi profiter de la symétrie de la ligne mais je n'ai pas cherché plus que ça a descendre la complexité

```python
def pascal(n):
    N = n-1
    p = fact(N)
    return [p/(fact(i)*fact(N-i)) for i in range(N+1)]
print(pascal(1))
print(pascal(2))
print(pascal(6))
```

b)Définissons une fonction `mdlast(lst,val)` ayant pour argument une liste de tuples d'entiers `lst` et un entier `val` et retourner la liste de tuples avec le dernier élément de chaque tuple remplacé par `val`. Par exemple si `lst = [(10, 20), (30, 40, 50, 60), (70, 80, 90)]` alors `mdlast(lst,100)` doit retourner `[(10, 100), (30, 40, 50, 100), (70, 80, 100)]`.

```python
def mdlast(lst, val):
    newLst=[]
    for i in range(len(lst)):
        lstProv=[]
        for j in range(len(lst[i])-1):
            lstProv.append(lst[i][j])
        lstProv.append(val)
        newLst.append(tuple(lstProv))
    return newLst
lst = [(10, 20), (30, 40, 50, 60), (70, 80, 90)]
print(mdlast(lst,100))
```

Commentaire : Ceette méthode fait un peuusine a gaz, en convertissant systématiquement les tuples en listes pour recréer des tuples.

a)Définir une fonction `divisible(n)` qui retourne l'ensemble des nombres entiers divisibles par `n` inférieurs ou égaux à 100.

```python
def divisible(N):
    return {N*i for i in range(100//N +1)}
print(divisible(7))
print(divisible(10))
```
